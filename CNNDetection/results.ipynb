### **A simple matplotlib program to print the results in a bargraph**

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 8)) 
X = ['progan', 'biggan', 'cyclegan', 'stargan', 'gaugan',
     'crn', 'imle', 'seeingdark', 'san', 'deepfake']
###### These are hard coded values ######
res_101 = np.array([0.51225, 0.53625, 0.5280090840272521, 0.49899949974987495, 0.5512, 0.5356471325603259,
           0.5704324663115011, 0.5166666666666667, 0.4931506849315068, 0.5013876040703052])*100

res_152 = np.array([0.52675, 0.5505, 0.5246025738077215, 0.5002501250625313, 0.5606, 0.5269507991225322,
           0.554763397054215, 0.5888888888888889, 0.4977168949771689, 0.5008325624421832])*100
###############
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, res_101, 0.4, label='resnet_101')
plt.bar(X_axis + 0.2, res_152, 0.4, label='resnet_152')

plt.xticks(X_axis, X)
plt.yticks(np.arange(0, 101,step=10))
plt.xlabel("Testing on different GANs",fontsize=15)
plt.ylabel("Accuracy",fontsize=15)
plt.title("Accuracy comparision between resnet_101 and resnet_152 classifier models",fontsize=15)
plt.legend(prop={"size":20})
plt.show()
```
